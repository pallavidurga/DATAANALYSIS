```python
sal = np.array([12000,60000,80000,76000])
wd = np.array([210,300,230,256])
years = [1,2]
bonus=[]
for i,j in zip(sal,wd):
 if j > 250:
    i = i+i*0.1
    bonus.append(i)
 else:
    i = i-i*0.1
    bonus.append(i)
print(bonus)
```

```python
import matplotlib.pyplot as plt
a = [230,560,780,127,128]
b = [200,160,270,127,400]
years = [1,2,3,4]
profit_a = [(a[i]-a[i-1]) for i in range(1,len(a))]
profit_b = [(b[i]-b[i-1]) for i in range(1,len(b))]
plt.subplot(2,1,2)
plt.plot(years,profit_a,color='hotpink',linewidth = '7',label ='CompanyA',marker='.',ms='20',mec='k')
plt.subplot(2,1,1)
plt.plot(years,profit_b,color='black',linestyle='dotted',label ='CompanyB',marker = 'H')
plt.show()
```
